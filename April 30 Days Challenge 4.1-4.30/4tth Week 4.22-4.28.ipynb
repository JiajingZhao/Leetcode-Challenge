# 22. Subarray Sum Equals K

Given an array of integers and an integer k, you need to find the total number of continuous subarrays whose sum equals to k.

Example 1:  
Input:nums = [1,1,1], k = 2  
Output: 2  
Note:  
The length of the array is in range [1, 20,000].  
The range of numbers in the array is [-1000, 1000] and the range of the integer k is [-1e7, 1e7].


** My Solution 暴力查找 超时**

```python
def subarraySum(nums, k) -> int:
    count = 0 
    length = len(nums)
    for i in range(length):
        for j in range(i+1,length+1):
            if sum(nums[i:j]) == k:
                count += 1
    return count
```

```python
def subarraySum2(nums, k) -> int:
    preSumCount={}
    curSum=0
    result=0
    preSumCount[0]=1
    for num in nums:
        curSum+=num
        if curSum-k in preSumCount:
            result+=preSumCount[curSum-k]
        preSumCount[curSum]=preSumCount.get(curSum,0)+1
    return result
```

```python
def subarraySum3(nums, k):
    """
    :type nums: List[int]
    :type k: int
    :rtype: int
    """
    count = 0
    sums = 0
    d = dict() #create a hash list
    d[0] = 1   # initialization, count the situation that single number = k

    for i in range(len(nums)):
        sums += nums[i]
        count += d.get(sums-k,0)    #times of (current sum-k) in dict
        d[sums] = d.get(sums,0) + 1 #dict record the sum and correponding times it shows up, add new sums each time
        print(count,sums,d)
    return count
```

# 23. Bitwise AND of Numbers Range

Given a range [m, n] where 0 <= m <= n <= 2147483647, return the bitwise AND of all numbers in this range, inclusive.

Example 1:   
Input: [5,7]  
Output: 4  

Example 2:  
Input: [0,1]  
Output: 0

**My Solution - immer time exceeded**  
**bitwise AND = "&", 1 when both 1,else 0**  
**bitwise OR = "|", 1 when there is 1**  
**bitewise XOR "^", 1 when different**

```python
def rangeBitwiseAnd(m, n) -> int:
    i = m
    while i >= m and i<=n:
        m = m&i
        i += 1
    return m  
```

***Solution2 - only related to the left common bits of m and n*

```python
def rangeBitwiseAnd2(m, n) -> int: 
    i = 0
    while m != n:
        m >>= 1 #right shift one bit,high bit complementing by 0
        n >>= 1
        i += 1
    return m << i # at the moment, m is the left commen part of m and n

# Runtime: 64 ms
# Memory Usage: 13.9 MB
```

```python
def rangeBitwiseAnd3(m, n) -> int: 
    while m < n:
        n &= (n - 1)
    return n
```

# 24. Least Recently Used (LRU) cache - Design Structure

Design and implement a data structure for Least Recently Used (LRU) cache. It should support the following operations: get and put.

**get(key)** - Get the value (will always be positive) of the key if the key exists in the cache, otherwise return -1.
**put(key, value)** - Set or insert the value if the key is not already present. When the cache reached its capacity, it should invalidate the least recently used item before inserting a new item. **达到capacity，踢出最不常使用的key**

The cache is initialized with a positive capacity.

Follow up:  Could you do both operations in **O(1)** time complexity?  

Example:  
LRUCache cache = new LRUCache( 2 /* capacity */ );

cache.put(1, 1);  
cache.put(2, 2); 
cache.get(1);       // returns 1  
cache.put(3, 3);    // evicts key 2  
cache.get(2);       // returns -1 (not found)  
cache.put(4, 4);    // evicts key 1  
cache.get(1);       // returns -1 (not found)  
cache.get(3);       // returns 3  
cache.get(4);       // returns 4  

**Requirement:   
Access to a random key in O(1) → hashtable   
Remove the last entry in LRU cache in O(1) → list   
Add/move an entry to the front of LRU cache in O(1) → list  
双向链表**

```python
class LRUCache: # [LRU >> RU]

    def __init__(self, capacity: int):
        self.deque = collections.deque([]) #双向队列
        self.dic = {} #solve prob of finding val of the key
        self.capacity = capacity        

    def get(self, key: int) -> int:
        if key not in self.dic: #if key not in dic, return -1
            return -1
        self.deque.remove(key) #1. move the key to the last loc of the deque.
        self.deque.append(key) 
        return self.dic[key]  #2. return val of the key   

    def put(self, key: int, value: int) -> None:
        if key in self.dic:    
            self.deque.remove(key)
        elif len(self.dic) == self.capacity: #if voll capacity
            v = self.deque.popleft()  #remove the Least Recently Used element
            self.dic.pop(v)
        self.deque.append(key)
        self.dic[key] = value    
```

# 25. Jump Game

Given an array of non-negative integers, you are initially positioned at the first index of the array.  
Each element in the array represents your **maximum jump length** at that position.  
Determine if you are able to reach the last index.

Example 1:

Input: nums = [2,3,1,1,4]  
Output: true  
Explanation: Jump 1 step from index 0 to 1, then 3 steps to the last index.

Example 2:

Input: nums = [3,2,1,0,4]  
Output: false  
Explanation: You will always arrive at index 3 no matter what. Its maximum jump length is 0, which makes it impossible to reach the last index.

**My Falsch Soution - not consider maximal length jump** 

```python
def canJump(nums) -> bool:
    i = 0
    N = len(nums)
    while i < N-1:
        if nums[i] == 0:
            return False
        i = i+nums[i]
    print("i:",i)
    if i >= N-1:
        return True
    else: 
        return False
```

**Solution2 - 贪心算法**

```python
def canJump2(nums) -> bool:
    far = nums[0] #far: maximal jump to "far"
    for i in range(len(nums)):
        if far < i:
            # it means we cannot reach i, break
            return False
        print("i:",i,"far:",far)
        far = max(far, i + nums[i])
    return far >= len(nums)-1
# Runtime: 84 ms
# Memory Usage: 15.8 MB
```

```python
nums1 = [2,3,1,1,4]
nums2 = [3,2,1,0,4]
nums3 = [2,0]
nums4 = [2,5,0,0]
print("<1>:",canJump2(nums1))
print("<2>:",canJump2(nums2))
print("<3>:",canJump2(nums3))
print("<4>:",canJump2(nums4))
```
